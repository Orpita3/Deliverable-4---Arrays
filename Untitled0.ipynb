```python
#@title Default title text
# Step (1): Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.drop(iris.columns[4], axis=1, inplace=True)
```

```python
iris.head()
```

```python
iris = iris.assign(sepal_sum = lambda x: x.sepal_length + x.sepal_width)
iris = iris.assign(petal_sum = lambda x: x.petal_length + x.petal_width)
```

```python
iris.head()
```

```python
data = {'avg': [], 'stddev': [], 'minm': [], 'maxm': []}
index = []
for col in iris.columns:
    index.append(col)
    data['avg'].append(round(iris[col].mean(),6))
    data['stddev'].append(round(iris[col].std(),6))
    data['minm'].append(round(iris[col].min(),6))
    data['maxm'].append(round(iris[col].max(),6))


index.insert(2,index[-2])
index.pop(-2)

for key in data:
    data[key].insert(2,data[key][-2])
    data[key].pop(-2)

new_df = pd.DataFrame(data, index)
```
